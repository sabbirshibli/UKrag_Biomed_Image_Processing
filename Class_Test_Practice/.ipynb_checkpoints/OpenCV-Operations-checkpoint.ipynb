```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('image.jpg')
```

```python
cv2.imshow("Original Image", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(image)
plt.show()
```

# Changing Colorspace

```python
# Converting BGR to RGB
img_bgr = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(2,2))
plt.imshow(img_bgr)
plt.show()
```

```python
# Converting RGB to Grey
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(2,2))
plt.imshow(img_gray, cmap='gray')
plt.show()
```

# Geometric Trasformation
### Learn to apply different geometric transformations to images like translation, rotation, affine transformation etc.

##### OpenCV provides two transformation functions, cv2.warpAffine and cv2.warpPerspective, with which you can have all kinds of transformations. cv2.warpAffine takes a 2x3 transformation matrix while cv2.warpPerspective takes a 3x3 transformation matrix as input.

## Scaling

```python
resized = cv2.resize(image, dsize=None, fx=.5,fy=.5, interpolation=cv2.INTER_NEAREST)
```

```python
plt.subplot(121)
plt.title("Original Image")
plt.imshow(image)
plt.subplot(122)
plt.title("Resized Image")
plt.imshow(resized)
plt.plot()
```

## Translation
#### Shifting Object Location

```python
rows, cols = img_gray.shape
```

```python
print(rows)
print(cols)
```

```python
M = np.float32([[1,0,100],[0,1,100]])
```

```python
trans = cv2.warpAffine(img_gray,M,(rows,cols))
```

```python
plt.subplot(121)
plt.title("Original Image")
plt.imshow(img_gray)
plt.subplot(122)
plt.title("Shifted Image")
plt.imshow(trans)
plt.plot()
```

## Rotation

```python
M = cv2.getRotationMatrix2D((rows/2, cols/2),45,1)
rotated = cv2.warpAffine(img_gray, M, (rows,cols))
```

```python
plt.subplot(121)
plt.title("Original Image")
plt.imshow(img_gray)
plt.subplot(122)
plt.title("Rotated Image")
plt.imshow(rotated)
plt.plot()
```

## Image Thresholding
### There are some types of thresholding
#### 1. Simple Thresholding
#### 2. Adaptive Thresholding
#### 3. Otsu's Thresholding

### Simple Thresholding

```python
# Gaussian Blur for better thresholding
blurred = cv2.GaussianBlur(img_gray,(5,5),0)
# Binary Thresholding
(T,thresh_bin) = cv2.threshold(blurred, 100, 256, cv2.THRESH_BINARY)
# TRUNC Thresholding
(T,thresh_trunc) = cv2.threshold(blurred, 100, 256, cv2.THRESH_TRUNC)
# TOZERO Thresholding
(T,thresh_tozero) = cv2.threshold(blurred, 100, 256, cv2.THRESH_TOZERO)
```

```python
plt.subplot(121)
plt.title("Original Image")
plt.imshow(img_gray, cmap='gray')
plt.subplot(122)
plt.title("Thresholded Image")
plt.imshow(thresh_tozero, cmap='gray')
plt.plot()
```

### Adaptive Thresholding

```python
thresh_ada = cv2.adaptiveThreshold(blurred,256,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,95,-10)
```

```python
plt.subplot(121)
plt.title("Original Image")
plt.imshow(img_gray, cmap='gray')
plt.subplot(122)
plt.title("Thresholded Image")
plt.imshow(thresh_ada, cmap='gray')
plt.plot()
```

### Otsu's Binarization

```python
(T, thresh_otsu) = cv2.threshold(blurred,100,2562,cv2.THRESH_OTSU|cv2.THRESH_BINARY)
```

```python
plt.subplot(121)
plt.title("Original Image")
plt.imshow(img_gray, cmap='gray')
plt.subplot(122)
plt.title("Thresholded Image")
plt.imshow(thresh_otsu, cmap='gray')
plt.plot()
```

# Image Smoothing (Blurring)
### There are few types of smoothing techniques:
#### 1. Normal Blur
#### 2. Gaussian Blur
#### 3. Median Blur
#### 4. Biletaral Blur

### Normal Blur

```python
# input a new noisy image to distinguish blur
img = cv2.imread('img.jpg',0)
```

```python
smooth = cv2.blur(img,(5,5))
```

```python
plt.subplot(121)
plt.title("Original Image")
plt.imshow(img, cmap='gray')
plt.subplot(122)
plt.title("Smooth Image")
plt.imshow(smooth, cmap='gray')
plt.plot()
```

### Gaussian Blur

```python
g_smooth = cv2.GaussianBlur(img,(5,5),0)
```

```python
plt.subplot(121)
plt.title("Original Image")
plt.imshow(img, cmap='gray')
plt.subplot(122)
plt.title("Smooth Image")
plt.imshow(g_smooth, cmap='gray')
plt.plot()
```

### Median Blur

```python
m_smooth = cv2.medianBlur(img, 5)
```

```python
plt.subplot(121)
plt.title("Original Image")
plt.imshow(img, cmap='gray')
plt.subplot(122)
plt.title("Smooth Image")
plt.imshow(m_smooth, cmap='gray')
plt.plot()
```

### Bilateral Blur

```python
b_smooth = cv2.bilateralFilter(img,80,75,75)
```

```python
plt.subplot(121)
plt.title("Original Image")
plt.imshow(img, cmap='gray')
plt.subplot(122)
plt.title("Smooth Image")
plt.imshow(b_smooth, cmap='gray')
plt.plot()
```

# Morphological Operations

### Erosion
#### Makes mask shape thinner

```python
# Importing a new image with mask
mask = cv2.imread('character.png',0)
```

```python
kernel = np.ones((5,5), np.uint8)
```

```python
erode = cv2.morphologyEx(mask,cv2.MORPH_ERODE,kernel,iterations=5)
```

```python
plt.subplot(121)
plt.title("Original Image")
plt.imshow(mask, cmap='gray')
plt.subplot(122)
plt.title("Erodated Image")
plt.imshow(erode, cmap='gray')
plt.plot()
```

### Dilation
#### Makes mask thicker

```python
dilated = cv2.morphologyEx(erode,cv2.MORPH_DILATE,kernel,iterations=5)
```

```python
plt.subplot(121)
plt.title("Original Image")
plt.imshow(mask, cmap='gray')
plt.subplot(122)
plt.title("Dilated Image")
plt.imshow(dilated, cmap='gray')
plt.plot()
```

### Opening
#### Erosion -> Dilation

```python
opening = cv2.morphologyEx(mask,cv2.MORPH_OPEN,kernel,iterations=5)
```

```python
plt.subplot(121)
plt.title("Original Image")
plt.imshow(mask, cmap='gray')
plt.subplot(122)
plt.title("Opened Image")
plt.imshow(opening, cmap='gray')
plt.plot()
```

### Closing
#### Dilation -> Erosion

```python
closing = cv2.morphologyEx(mask,cv2.MORPH_CLOSE,kernel,iterations=5)
```

```python
plt.subplot(121)
plt.title("Original Image")
plt.imshow(mask, cmap='gray')
plt.subplot(122)
plt.title("Closed Image")
plt.imshow(closing, cmap='gray')
plt.plot()
```

```python
### Gradiant
gradiant = cv2.morphologyEx(mask,cv2.MORPH_GRADIENT,kernel)
```

```python
plt.subplot(121)
plt.title("Original Image")
plt.imshow(mask, cmap='gray')
plt.subplot(122)
plt.title("Gradiant Image")
plt.imshow(gradiant, cmap='gray')
plt.plot()
```
